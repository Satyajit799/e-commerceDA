# Correlation Analysis

## Objective

Analyze the relationships among key business metrics using a correlation matrix and heatmap.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../datas/cleaned_ecommerce.csv")
```

```python
correlation_data = df[[
    "Sales",
    "Quantity",
    "Discount",
    "Profit",
    "Shipping Cost",
    "Aging"
]]

correlation_data.head()
```

```python
correlation_matrix = correlation_data.corr()

correlation_matrix
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap="coolwarm",
    linewidths=0.5,
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.tight_layout()

plt.savefig("../charts/correlation_heatmap.png", dpi=300)

plt.show()
```

```python
profit_corr = correlation_matrix["Profit"].sort_values(ascending=False)

profit_corr
```

```python
sales_corr = correlation_matrix["Sales"].sort_values(ascending=False)

sales_corr
```

```python
print("Correlation with Profit")
print(profit_corr)

print("\n")

print("Correlation with Sales")
print(sales_corr)
```

```python
sns.pairplot(
    correlation_data
)

plt.savefig("../charts/pairplot.png", dpi=300)

plt.show()
```

- Identified variables most strongly associated with sales and profit.
