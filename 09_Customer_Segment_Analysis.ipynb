# Customer Segment Analysis

## Objective

Analyze the performance of each customer segment based on:
- Total Sales
- Total Profit
- Average Order Value (AOV)

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../datas/cleaned_ecommerce.csv")
```

```python
segment_analysis = (
    df.groupby("Segment")
      .agg(
          Total_Sales=("Sales", "sum"),
          Total_Profit=("Profit", "sum"),
          Total_Orders=("Order ID", "nunique")
      )
)

segment_analysis["Average_Order_Value"] = (
    segment_analysis["Total_Sales"] /
    segment_analysis["Total_Orders"]
)

segment_analysis
```

```python
print("Highest Sales Segment :")
print(segment_analysis["Total_Sales"].idxmax())

print()

print("Highest Profit Segment :")
print(segment_analysis["Total_Profit"].idxmax())

print()

print("Highest Average Order Value :")
print(segment_analysis["Average_Order_Value"].idxmax())
```

```python
plt.figure(figsize=(8,5))

bars = plt.bar(
    segment_analysis.index,
    segment_analysis["Total_Sales"]
)

plt.title("Total Sales by Customer Segment")
plt.xlabel("Customer Segment")
plt.ylabel("Sales ($)")

for bar in bars:
    plt.text(
        bar.get_x()+bar.get_width()/2,
        bar.get_height(),
        f"${bar.get_height():,.0f}",
        ha="center",
        va="bottom"
    )

plt.tight_layout()

plt.savefig("../charts/customer_segment_sales.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(8,5))

bars = plt.bar(
    segment_analysis.index,
    segment_analysis["Total_Profit"]
)

plt.title("Total Profit by Customer Segment")
plt.xlabel("Customer Segment")
plt.ylabel("Profit ($)")

for bar in bars:
    plt.text(
        bar.get_x()+bar.get_width()/2,
        bar.get_height(),
        f"${bar.get_height():,.0f}",
        ha="center",
        va="bottom"
    )

plt.tight_layout()

plt.savefig("../charts/customer_segment_profit.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(8,5))

bars = plt.bar(
    segment_analysis.index,
    segment_analysis["Average_Order_Value"]
)

plt.title("Average Order Value by Customer Segment")
plt.xlabel("Customer Segment")
plt.ylabel("Average Order Value ($)")

for bar in bars:
    plt.text(
        bar.get_x()+bar.get_width()/2,
        bar.get_height(),
        f"${bar.get_height():.2f}",
        ha="center",
        va="bottom"
    )

plt.tight_layout()

plt.savefig("../charts/customer_segment_aov.png", dpi=300)

plt.show()
```

```python
segment_analysis[[
    "Total_Sales",
    "Total_Profit"
]].plot(
    kind="bar",
    figsize=(10,6)
)

plt.title("Customer Segment Performance")
plt.xlabel("Customer Segment")
plt.ylabel("Amount ($)")

plt.xticks(rotation=0)

plt.tight_layout()

plt.savefig("../charts/customer_segment_performance.png", dpi=300)

plt.show()
```
