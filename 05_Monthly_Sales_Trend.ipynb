```python
# Monthly Sales Trend Analysis

## Objective

# Analyze monthly sales performance to identify seasonal trends, peak sales months, and low-performing months.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../datas/cleaned_ecommerce.csv")
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
df["Year_Month"] = df["Order Date"].dt.to_period("M")
```

```python
monthly_sales = (
    df.groupby("Year_Month")["Sales"]
      .sum()
)

monthly_sales
```

```python
print("Highest Sales Month")
print(monthly_sales.idxmax())
print(monthly_sales.max())

print()

print("Lowest Sales Month")
print(monthly_sales.idxmin())
print(monthly_sales.min())
```

```python
plt.figure(figsize=(14,6))

plt.plot(
    monthly_sales.index.astype(str),
    monthly_sales.values,
    marker="o",
    linewidth=2
)

plt.title("Monthly Sales Trend")

plt.xlabel("Month")

plt.ylabel("Sales ($)")

plt.xticks(rotation=45)

plt.grid(True)

plt.tight_layout()

plt.savefig("../charts/monthly_sales_trend.png", dpi=300)

plt.show()
```

```python
highest_month = monthly_sales.idxmax()

lowest_month = monthly_sales.idxmin()

print(f"Highest Sales Month : {highest_month}")

print(f"Lowest Sales Month : {lowest_month}")
```

```python
average_sales = monthly_sales.mean()

print(f"Average Monthly Sales : ${average_sales:,.2f}")
```
