```python
# Top 10 Best-Selling Products

# This notebook identifies the top 10 products based on total sales and visualizes the results using a horizontal bar chart.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../datas/cleaned_ecommerce.csv")
```

```python
df.head()
```

```python
top_products = (
    df.groupby("Product")["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_products
```

```python
plt.figure(figsize=(12,6))

plt.barh(
    top_products.index,
    top_products.values
)

plt.title("Top 10 Best-Selling Products")

plt.xlabel("Total Sales")

plt.ylabel("Product")

plt.gca().invert_yaxis()

plt.show()
```

```python
plt.figure(figsize=(12,6))

bars = plt.barh(
    top_products.index,
    top_products.values
)

plt.title("Top 10 Best-Selling Products")

plt.xlabel("Total Sales")

plt.ylabel("Product")

plt.gca().invert_yaxis()

for bar in bars:
    plt.text(
        bar.get_width(),
        bar.get_y() + bar.get_height()/2,
        f"${bar.get_width():,.0f}",
        va="center"
    )

plt.show()
```

```python
plt.savefig("../charts/top10_products.png", dpi=300, bbox_inches="tight")
```
