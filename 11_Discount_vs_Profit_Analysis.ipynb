# Discount vs Profit Analysis

## Objective

 Analyze the relationship between discount and profit to determine whether higher discounts reduce profitability.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../datas/cleaned_ecommerce.csv")
```

```python
df[["Discount", "Profit"]].head()
```

```python
correlation = df["Discount"].corr(df["Profit"])

print(f"Correlation between Discount and Profit: {correlation:.4f}")
```

```python
plt.figure(figsize=(10,6))

sns.regplot(
    data=df,
    x="Discount",
    y="Profit",
    scatter_kws={"alpha":0.5},
    line_kws={"color":"red"}
)

plt.title("Discount vs Profit")
plt.xlabel("Discount")
plt.ylabel("Profit ($)")

plt.tight_layout()

plt.savefig("../charts/discount_vs_profit.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(5,4))

sns.heatmap(
    df[["Discount", "Profit"]].corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Matrix")

plt.tight_layout()

plt.savefig("../charts/discount_profit_heatmap.png", dpi=300)

plt.show()
```

```python
discount_summary = (
    df.groupby("Discount")
      .agg(
          Average_Profit=("Profit", "mean"),
          Total_Sales=("Sales", "sum"),
          Orders=("Order ID", "count")
      )
      .reset_index()
)

discount_summary
```

## Business Insights

- Calculated the correlation between discount and profit.
- Visualized the relationship using a scatter plot with a regression line.
- A negative correlation suggests that increasing discounts may reduce profitability.
- Businesses should optimize discount strategies to balance customer acquisition and profit margins.
