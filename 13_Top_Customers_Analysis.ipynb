# Top Customers Analysis

## Objective

Identify the top 20 customers based on:
-Total Sales
- Total Profit

This analysis helps identify the company's most valuable customers.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../datas/cleaned_ecommerce.csv")
```

```python
top_sales_customers = (
    df.groupby("Customer Name")
      .agg(
          Total_Sales=("Sales", "sum")
      )
      .sort_values(by="Total_Sales", ascending=False)
      .head(20)
)

top_sales_customers
```

```python
top_profit_customers = (
    df.groupby("Customer Name")
      .agg(
          Total_Profit=("Profit", "sum")
      )
      .sort_values(by="Total_Profit", ascending=False)
      .head(20)
)

top_profit_customers
```

```python
print("Top Customer by Sales")

print("----------------------------")

print(top_sales_customers.head(1))
```

```python
print("Top Customer by Profit")

print("----------------------------")

print(top_profit_customers.head(1))
```

```python
plt.figure(figsize=(12,8))

bars = plt.barh(
    top_sales_customers.index,
    top_sales_customers["Total_Sales"]
)

plt.title("Top 20 Customers by Sales")

plt.xlabel("Sales ($)")
plt.ylabel("Customer Name")

plt.gca().invert_yaxis()

for bar in bars:
    plt.text(
        bar.get_width(),
        bar.get_y() + bar.get_height()/2,
        f"${bar.get_width():,.0f}",
        va="center"
    )

plt.tight_layout()

plt.savefig("../charts/top20_customers_sales.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(12,8))

bars = plt.barh(
    top_profit_customers.index,
    top_profit_customers["Total_Profit"]
)

plt.title("Top 20 Customers by Profit")

plt.xlabel("Profit ($)")
plt.ylabel("Customer Name")

plt.gca().invert_yaxis()

for bar in bars:
    plt.text(
        bar.get_width(),
        bar.get_y() + bar.get_height()/2,
        f"${bar.get_width():,.0f}",
        va="center"
    )

plt.tight_layout()

plt.savefig("../charts/top20_customers_profit.png", dpi=300)

plt.show()
```

```python
customer_summary = (
    df.groupby("Customer Name")
      .agg(
          Total_Sales=("Sales", "sum"),
          Total_Profit=("Profit", "sum"),
          Orders=("Order ID", "nunique")
      )
      .sort_values(by="Total_Sales", ascending=False)
      .head(20)
)

customer_summary
```
