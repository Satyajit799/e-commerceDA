```python
# Product Category Analysis

## Objective

#Analyze the performance of each product category based on:
#- Total Sales
#- Total Profit
#- Total Quantity Sold
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../datas/cleaned_ecommerce.csv")
```

```python
category_analysis = (
    df.groupby("Product Category")
      .agg({
          "Sales":"sum",
          "Profit":"sum",
          "Quantity":"sum"
      })
      .sort_values(by="Sales", ascending=False)
)

category_analysis
```

```python
print("Highest Sales Category:")
print(category_analysis["Sales"].idxmax())

print()

print("Highest Profit Category:")
print(category_analysis["Profit"].idxmax())

print()

print("Highest Quantity Sold Category:")
print(category_analysis["Quantity"].idxmax())
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    category_analysis.index,
    category_analysis["Sales"]
)

plt.title("Sales by Product Category")

plt.xlabel("Category")

plt.ylabel("Sales ($)")

plt.tight_layout()

plt.savefig("../charts/category_sales.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    category_analysis.index,
    category_analysis["Profit"]
)

plt.title("Profit by Product Category")

plt.xlabel("Category")

plt.ylabel("Profit ($)")

plt.tight_layout()

plt.savefig("../charts/category_profit.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    category_analysis.index,
    category_analysis["Quantity"]
)

plt.title("Quantity Sold by Product Category")

plt.xlabel("Category")

plt.ylabel("Quantity")

plt.tight_layout()

plt.savefig("../charts/category_quantity.png", dpi=300)

plt.show()
```

```python
category_analysis.plot(
    kind="bar",
    figsize=(10,6)
)

plt.title("Product Category Analysis")

plt.xlabel("Product Category")

plt.ylabel("Values")

plt.xticks(rotation=0)

plt.tight_layout()

plt.savefig("../charts/product_category_analysis.png", dpi=300)

plt.show()
```
