# Profit Distribution by Product Category

## Objective

Analyze the distribution of profit across different product categories using boxplots and violin plots to identify variability, outliers, and profitability patterns.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../datas/cleaned_ecommerce.csv")
```

```python
profit_summary = (
    df.groupby("Product Category")["Profit"]
      .describe()
)

profit_summary
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    data=df,
    x="Product Category",
    y="Profit"
)

plt.title("Profit Distribution by Product Category")

plt.xlabel("Product Category")
plt.ylabel("Profit ($)")

plt.tight_layout()

plt.savefig("../charts/profit_boxplot_category.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.violinplot(
    data=df,
    x="Product Category",
    y="Profit"
)

plt.title("Profit Distribution by Product Category (Violin Plot)")

plt.xlabel("Product Category")
plt.ylabel("Profit ($)")

plt.tight_layout()

plt.savefig("../charts/profit_violin_category.png", dpi=300)

plt.show()
```

```python
avg_profit = (
    df.groupby("Product Category")["Profit"]
      .mean()
      .sort_values(ascending=False)
)

avg_profit
```

```python
plt.figure(figsize=(8,5))

bars = plt.bar(
    avg_profit.index,
    avg_profit.values
)

plt.title("Average Profit by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Average Profit ($)")

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        f"${bar.get_height():.2f}",
        ha="center",
        va="bottom"
    )

plt.tight_layout()

plt.savefig("../charts/average_profit_category.png", dpi=300)

plt.show()
```

```python
highest = avg_profit.idxmax()
lowest = avg_profit.idxmin()

print(f"Highest Average Profit Category : {highest}")
print(f"Lowest Average Profit Category  : {lowest}")
```

## Business Insights

- Compared profit distributions across product categories.
- Identified categories with high variability in profit.
- Detected outliers using the boxplot.
- Used the violin plot to understand the density and spread of profits.
- Calculated average profit for each category.

### Recommendations

- Focus on categories with consistently high profitability.
- Investigate categories showing large negative outliers.
- Optimize pricing and promotional strategies for low-profit categories.
