```python
# Overall Sales Performance

#In this notebook, we calculate the key business performance indicators (KPIs) for the e-commerce dataset.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../datas/cleaned_ecommerce.csv")
```

```python
total_sales = df["Sales"].sum()

total_profit = df["Profit"].sum()

total_quantity = df["Quantity"].sum()

total_shipping_cost = df["Shipping Cost"].sum()

total_orders = df["Order ID"].nunique()

total_customers = df["Customer ID"].nunique()

average_order_value = total_sales / total_orders

profit_margin = (total_profit / total_sales) * 100
```

```python
print("=" * 50)
print("OVERALL SALES PERFORMANCE")
print("=" * 50)

print(f"Total Sales          : ${total_sales:,.2f}")
print(f"Total Profit         : ${total_profit:,.2f}")
print(f"Total Quantity Sold  : {total_quantity:,.0f}")
print(f"Shipping Cost        : ${total_shipping_cost:,.2f}")
print(f"Total Orders         : {total_orders}")
print(f"Total Customers      : {total_customers}")
print(f"Average Order Value  : ${average_order_value:,.2f}")
print(f"Profit Margin        : {profit_margin:.2f}%")
```

```python
kpi_names = [
    "Sales",
    "Profit",
    "Shipping Cost"
]

kpi_values = [
    total_sales,
    total_profit,
    total_shipping_cost
]

plt.figure(figsize=(8,5))

plt.bar(kpi_names, kpi_values)

plt.title("Overall Sales Performance")

plt.ylabel("Amount")

plt.savefig("../charts/overall_business_performance.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.bar(
    ["Sales", "Profit"],
    [total_sales, total_profit]
)

plt.title("Sales vs Profit")
plt.ylabel("Amount ($)")
plt.savefig("../charts/sales_vs_profit.png", dpi=300)

plt.show()

plt.figure(figsize=(7,5))

plt.bar(
    ["Orders", "Customers"],
    [total_orders, total_customers]
)

plt.title("Orders vs Customers")
plt.ylabel("Count")

plt.show()

plt.figure(figsize=(6,5))

plt.bar(
    ["Quantity Sold"],
    [total_quantity]
)

plt.title("Total Quantity Sold")

plt.ylabel("Units")
plt.savefig("../charts/quantity_sold.png", dpi=300)

plt.show()

plt.figure(figsize=(6,6))

plt.pie(
    [profit_margin, 100-profit_margin],
    labels=["Profit Margin","Remaining"],
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Profit Margin")
plt.savefig("../charts/profit_margin.png", dpi=300)

plt.show()

plt.figure(figsize=(6,5))

plt.bar(
    ["Average Order Value"],
    [average_order_value]
)

plt.title("Average Order Value")

plt.ylabel("$")

plt.savefig("../charts/average_order_value.png", dpi=300)
plt.show()
```
