# Ship Mode Performance Analysis

## Objective

Analyze the performance of each shipping mode using:
- Average Shipping Cost
- Average Delivery Aging
- Total Sales

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../datas/cleaned_ecommerce.csv")
```

```python
ship_mode_analysis = (
    df.groupby("Ship Mode")
      .agg(
          Average_Shipping_Cost=("Shipping Cost", "mean"),
          Average_Delivery_Aging=("Aging", "mean"),
          Total_Sales=("Sales", "sum")
      )
      .sort_values(by="Total_Sales", ascending=False)
)

ship_mode_analysis
```

```python
plt.figure(figsize=(8,5))

bars = plt.bar(
    ship_mode_analysis.index,
    ship_mode_analysis["Average_Shipping_Cost"]
)

plt.title("Average Shipping Cost by Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Average Shipping Cost ($)")

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        f"${bar.get_height():.2f}",
        ha="center",
        va="bottom"
    )

plt.tight_layout()

plt.savefig("../charts/average_shipping_cost.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(8,5))

bars = plt.bar(
    ship_mode_analysis.index,
    ship_mode_analysis["Average_Delivery_Aging"]
)

plt.title("Average Delivery Aging by Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Average Delivery Days")

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        f"{bar.get_height():.1f}",
        ha="center",
        va="bottom"
    )

plt.tight_layout()

plt.savefig("../charts/average_delivery_aging.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(8,5))

bars = plt.bar(
    ship_mode_analysis.index,
    ship_mode_analysis["Total_Sales"]
)

plt.title("Total Sales by Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Sales ($)")

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        f"${bar.get_height():,.0f}",
        ha="center",
        va="bottom"
    )

plt.tight_layout()

plt.savefig("../charts/ship_mode_sales.png", dpi=300)

plt.show()
```

```python
ship_mode_analysis.plot(
    kind="bar",
    figsize=(10,6)
)

plt.title("Ship Mode Performance")
plt.xlabel("Ship Mode")
plt.ylabel("Values")

plt.xticks(rotation=0)

plt.tight_layout()

plt.savefig("../charts/ship_mode_performance.png", dpi=300)

plt.show()
```

```python
fastest_ship = ship_mode_analysis["Average_Delivery_Aging"].idxmin()

fastest_days = ship_mode_analysis["Average_Delivery_Aging"].min()

print(f"Fastest Shipping Mode : {fastest_ship}")

print(f"Average Delivery Time : {fastest_days:.2f} Days")
```

## Business Insights

- Identified the fastest shipping method based on delivery aging.
- Compared total sales generated through each shipping mode.
