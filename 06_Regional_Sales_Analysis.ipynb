```python
# Regional Sales Analysis

## Objective

# Analyze sales and profit across different regions to identify the highest-performing and most profitable region.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../datas/cleaned_ecommerce.csv")
```

```python
region_analysis = (
    df.groupby("Region")
      .agg({
          "Sales": "sum",
          "Profit": "sum"
      })
      .sort_values(by="Sales", ascending=False)
)

region_analysis
```

```python
highest_sales_region = region_analysis["Sales"].idxmax()
highest_profit_region = region_analysis["Profit"].idxmax()

print("Highest Sales Region :", highest_sales_region)
print("Highest Profit Region :", highest_profit_region)
```

```python
plt.figure(figsize=(12,5))

bars = plt.bar(
    region_analysis.index,
    region_analysis["Sales"]
)

plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales ($)")

for bar in bars:
    plt.text(
        bar.get_x()+bar.get_width()/2,
        bar.get_height(),
        f"${bar.get_height():,.0f}",
        ha="center",
        va="bottom"
    )

plt.tight_layout()

plt.savefig("../charts/region_sales.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(12,5))

bars = plt.bar(
    region_analysis.index,
    region_analysis["Profit"]
)

plt.title("Total Profit by Region")
plt.xlabel("Region")
plt.ylabel("Profit ($)")

for bar in bars:
    plt.text(
        bar.get_x()+bar.get_width()/2,
        bar.get_height(),
        f"${bar.get_height():,.0f}",
        ha="center",
        va="bottom"
    )

plt.tight_layout()

plt.savefig("../charts/region_profit.png", dpi=300)

plt.show()
```

```python
region_analysis.plot(
    kind="bar",
    figsize=(10,6)
)

plt.title("Regional Sales & Profit Comparison")
plt.xlabel("Region")
plt.ylabel("Amount ($)")

plt.xticks(rotation=0)

plt.tight_layout()

plt.savefig("../charts/regional_sales_profit.png", dpi=300)

plt.show()
```
